```python
x=np.array(["AIML","UI/UX","CWS","FLUTTER","CN"])
y=np.array([85,96,75,98,86])
myexplode=np.array([0,0,0,0.2,0])
col=np.array(["Red","c","Yellow","Green","Blue"])
plt.pie(y,labels=x,startangle=90,explode=myexplode,shadow=True,colors=col)
```

```python
x=np.array(["AIML","UI/UX","CWS","FLUTTER","CN"])
y=np.array([85,96,75,98,86])
myexplode=np.array([0,0,0,0.2,0])

def show_marks(pct):
    total=sum(y)
    marks=int(round(pct*total/100))
    return str(marks)

col=np.array(["Red","c","Yellow","Green","Blue"])
plt.pie(y,labels=x,startangle=90,explode=myexplode,shadow=True,colors=col,autopct=show_marks)
plt.legend(title="Subject Name")
```
