```python
x=nm.array([41,59,85])
y=nm.array([10,54,26])

print("X+Y",x+y)
print("X-Y",x-y)
print("X*Y",x*y)
print("X/Y",x/y)
```

```python
add=nm.add(x,y)
print("Subtraction :",nm.subtract(x,y))
print("Multiplication :",nm.multiply(x,y))
print("Division :",nm.divide(x,y))
print("modulo :",nm.mod(x,y))
```

```python
a=nm.array([10,20,30,40,50,52,36,441])
print("Max Value :",nm.max(a))
print("Min Value :",nm.min(a))
print("Sum Value :",nm.sum(a))
print("Mean Value :",nm.mean(a))
```
