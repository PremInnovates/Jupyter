```python
import numpy as np
```

```python
arr1=np.arange(2,21,2)
print(arr1)
```

```python
arr2=np.linspace(0,1,5)
print(arr2)
```

```python
arr3=np.full((3,3),7)
print(arr3)
```

```python
arr4=np.eye(4)
print(arr4)
```

```python
arr5=np.random.rand(10)
print(arr5)
```

```python
arr6=np.arange(1,13).reshape(3,4)
print(arr6)
```

```python
print(arr6.reshape)
print(arr6.ndim)
```

```python
arr8=np.arange(1,26).reshape(5,5)
print(arr8)
```

```python
first_row=arr8[0]
print(first_row)
```

```python
last_column=arr8[:,-1]
print(last_column)
```

```python
arr8[arr8>15]=0
print(arr8)
```
