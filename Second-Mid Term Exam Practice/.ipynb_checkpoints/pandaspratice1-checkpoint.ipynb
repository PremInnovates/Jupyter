```python
import pandas as pd
import numpy as mp
import matplotlib.pyplot as plt
import itertools
```

```python
df=pd.read_csv('customer.csv')
print("Top 10 Records :")
print(df.head(10))
```

```python
sorted_age=df.sort_values("Age")
print(sorted_age)
```

```python
df=df.dropna()
print("After Removing Empty Records")
print(df)
```

```python
plt.plot([df['Age'],df['Balance']])
plt.show()
```

```python
top_credited=df.sort_values("CreditScore",ascending=False).head(15)
print(top_credited)
```

```python
geo_counts=df["geography"].value_counts()
colors=["Red","Blue","Green","Orange","Purple"]
plt.figure()
plt.pie(geo_counts,label=geo_counts.index,colors=colors[:len(geo_counts)],autopct="1.1f%%")
plt.title("Customers By Geometry")
plt.show()
```
