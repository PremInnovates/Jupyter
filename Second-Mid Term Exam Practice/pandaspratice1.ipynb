```python
import pandas as pd
import numpy as mp
import matplotlib.pyplot as plt
import itertools
```

```python
df=pd.read_csv('customer.csv')
print("Top 10 Records :")
print(df.head(10))
```

```python
sorted_age=df.sort_values("Age")
print(sorted_age)
```

```python
df=df.dropna()
print("After Removing Empty Records")
print(df)
```

```python
plt.plot([df['Age'],df['Balance']])
plt.show()
```

```python
top_credited=df.sort_values("CreditScore",ascending=False).head(15)
print(top_credited)
```

```python
geo_counts=df["Geography"].value_counts()
colors=["Red","Blue","Green"]
plt.figure()
plt.pie(geo_counts,labels=geo_counts.index,colors=colors[:len(geo_counts)],autopct="%1.1f%%")
plt.title("Customer By Geography")
plt.show()
```

```python
high_salary=df[df["EstimatedSalary"]>75000]
print("\n Customer with Salary > 75000")
print(high_salary)
```

```python
mean_balanced=df["Balance"].mean()
print("\n Mean Balance :",mean_balanced)
```

```python
perm=list(itertools.permutations(df["CreditScore"].head(5)))
print(perm)
```

```python
plt.figure()
plt.scatter(df["Geography"],df["Balance"])
plt.xlabel("Geography")
plt.ylabel("Balance")
plt.title("Geography Vs Balance")
plt.show()
```

```python
#set=2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("pokemon_data.csv")
print(df.tail())
```

```python
df["Type2"]=df["Type2"].fillna("FILLED")
print("After Filling Empty Type2 :")
print(df)
```

```python
median_defense=df["Defense"].median()
print("median of Defence",median_defense)
```

```python
average_values=df[["HP","Attack","Defense"]].mean()
print("\n After HP,Attack,Defense :")
print(average_values)
```

```python
plt.figure()
plt.bar(df["Name"],df["Attack"])
plt.xticks(rotation=90)
plt.xlabel("Name")
plt.ylabel("Attack")
plt.title("Name Vs Attack")
plt.show()
```

```python
speed_sum=np.sum(df["Speed"])
print("\n Sum of Speed :",speed_sum)
```

```python
legendary=df[df["Legendary"]==True]
print("\n Legendary Pokemons :")
print(legendary)
```

```python
plt.figure()
plt.hist(df["Generation"])
plt.xlabel("Generation")
plt.ylabel("Count")
plt.title("generation Histogram")
plt.show()
```

```python
std_speed=df["Speed"].std()
print("\n Standard Deviation of Spped :",std_speed)
```

```python
plt.figure()
plt.scatter(df["HP"],df["Speed"])
plt.xlabel("HP")
plt.ylabel("Speed")
plt.title("HP Vs Speed")
plt.show()
```
