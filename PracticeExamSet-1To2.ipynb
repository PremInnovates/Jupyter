```python
import pandas as pd
import numpy as mp
import matplotlib.pyplot as plt
import itertools
```

```python
#load data and print top 10 records
df=pd.read_csv("customer.csv")
print("Top 10 Records :")
print(df.head(10))
```

```python
#sort data on age fields
sorted_age=df.sort_values("Age")
print(sorted_age)
```

```python
#remove empty records
df=df.dropna()
print("\n After Removing Empty Records")
print(df)
```

```python
#linechart on age and balance
# Age=df["Age"].tolist()
# Balance=df["Balance"].tolist()
# plt.plot(Age,Balance)
# plt.xlabel("Age")
# plt.ylabel("Balance")
# plt.title("Age Vs Balance")
# plt.show()

plt.plot([df['Age'],df['Balance']])
plt.show()
```

```python
#sort creditscore and display top 15 highest creditscore
top_credit=df.sort_values("CreditScore", ascending=False).head(15)
print(top_credit)
```

```python
#pie chart using geogrphy with customised colours
geo_counts=df["Geography"].value_counts()
colors=["Red","Blue","Green","Orange","Purple"]
plt.figure()
plt.pie(geo_counts,labels=geo_counts.index,colors=colors[:len(geo_counts)],
autopct="%1.1f%%")
plt.title("Customers By Geography")
plt.show()
```

```python
#customers with estimatedsalary
high_salary=df[df["EstimatedSalary"]>75000]
print("\n Customers with Salary > 75000")
print(high_salary)
```

```python
#Mean of Balance field
mean_balance=df["Balance"].mean()
print("\n Mean Balance :",mean_balance)
```

```python
#permutation 
perm=list(itertools.permutations(df["CreditScore"].head(5)))
print(perm)
```

```python
#scatter plot geography and balance
plt.figure()
plt.scatter(df["Geography"],df["Balance"])
plt.xlabel("Geography")
plt.ylabel("Balance")
plt.title("geography VS Balance")
plt.show()
```

```python
#set 2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("pokemon_data.csv")
print(df.tail())
```

```python
#fill empty type2 field with 'FILLED'
df["Type2"]=df["Type2"].fillna("FILLED")
print("\n After Filling Empty Type2 :")
print(df)
```

```python
#median of defence
median_defense=df["Defense"].median()
print("\n Median Of Defence :",median_defense)
```

```python
#average of HP, Attack, Defense
average_values=df[["HP","Attack","Defense"]].mean()
print("\n After of HP,Attack,Defense :")
print(average_values)
```

```python
#bar chart name and attack
plt.figure()
plt.bar(df["Name"],df["Attack"])
plt.xticks(rotation=90)
plt.xlabel("Name")
plt.ylabel("Attack")
plt.title("Name Vs Attack")
plt.show()
```

```python
speed_sum=np.sum(df["Speed"])
print("\nSum of Speed :",speed_sum)
```

```python
#Display Legency = True

legendary=df[df["Legendary"]==True]
print("\n Lengedary Pokemons :")
print(legendary)
```

```python
plt.figure()
plt.hist(df["Generation"])
plt.xlabel("Generation")
plt.ylabel("Count")
plt.title("Generation Histogram")
plt.show()
```

```python
#standard deviation of Speed
std_speed=df["Speed"].std()
print("\n Standard Deviation of Speed :",std_speed)
```

```python
plt.figure()
plt.scatter(df["HP"],df["Speed"])
plt.xlabel("HP")
plt.ylabel("Speed")
plt.title("HP vs Speed")
plt.show()
```
