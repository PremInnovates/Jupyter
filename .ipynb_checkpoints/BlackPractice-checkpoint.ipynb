```python
import numpy as np
arr1=np.arange(2,21,2) # print 2 to 20 only even numbers
print(arr1)
```

```python
arr2=np.linspace(0,1,5) #Array of 5 Equally spaced Number between 0 to 1
print(arr2)
```

```python
arr3=np.full((3,3),7) #3x3 array Filled With 7
print(arr3)
```

```python
arr4=np.eye(4)  #4x4 Identity Matrix
print(arr4)
```

```python
arr5=np.random.rand(10) #array of 10 random floating-point numbers between 0 and 1
print(arr5)
```

```python
arr6=np.arange(1,13).reshape(3,4) #reshape 1D Array number from 1 to 12 into a 3x4 Matrix
print(arr6)
```

```python
print(arr6.reshape) #Shape of Array
print(arr6.ndim) #shape of Dimension
```

```python
arr8=np.arange(1,26).reshape(5,5)  #array 1 to 25 reshape it into 5x5 matrix
print(arr8)
```

```python
first_row=arr8[0] #Extract First Row in Array
print(first_row)
```

```python
last_column=arr8[:,-1] #Extract last Column form array
print(last_column)
```

```python
arr8[arr8>15]=0 #replace all value greater than 15 in the array with 0
print(arr8)
```
