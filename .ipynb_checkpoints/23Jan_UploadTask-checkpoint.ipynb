```python
import matplotlib.pyplot as plt
```

```python
months=["Jan","Feb","Mar","Apr","May","Jun"]
sales=[12000,15000,18000,17000,21000,25000]
```

```python
plt.plot(months,sales)
plt.show()
```

```python
plt.plot(months,sales)
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
plt.plot(months,sales,linestyle="--",marker="o")
plt.show()
```

```python
plt.bar(months,sales)
```

```python
plt.pie(sales,labels=months,autopct="%1.1f%%")
plt.show()
```

```python
plt.scatter(range(len(months)),sales)
plt.show()
```

```python
plt.hist(sales,bins=5)
plt.show()
```

```python
profit=[2000,3000,4000,3500,4500,6000]
plt.plot(months,sales)
plt.plot(months,profit)
plt.show()
```

```python
plt.plot(months,sales)
plt.savefig("sales_chart.png")
plt.show()
```

```python
df=pd.read_csv("students.csv")
print(type(df))
```

```python
grouped=df.groupby("Course")["Marks"].sum()
print(grouped)
```

```python
grouped.plot(kind="bar")
plt.title("Total Marks By Course")
plt.xlabel=("Course")
plt.ylabel=("Marks")
plt.show()
```

```python
df["Marks"].plot(kind="hist")
plt.title("Marks Distribution")
plt.show()
```

```python
df.boxplot(column="Marks")
plt.title("Marks BoxPlot")
plt.show()
```

```python
df["Marks"].plot(kind="line")
plt.title("Marks Line Chart")
plt.show()
```

```python
df["Course"].value_counts().plot(
kind="pie",
autopct="%1.1f%%"
)
plt.ylabel("")
plt.show()
```
