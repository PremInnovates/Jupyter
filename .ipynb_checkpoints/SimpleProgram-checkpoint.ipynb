```python
import numpy as nm
```

```python
arr=nm.array([12,85,54,14,41])
print(arr)
```

```python
b=nm.array([[5,12,54],[41,256,52],[54,12,41]]) #MultiDimension Array
print(b)
```

```python
print("Dimension :",arr.ndim) #ndim used to display Dimension of An Array
```

```python
print(b.ndim)
```

```python
print("Shape :",b.shape) #Display Size OF Array
```

```python
print("Type :",b.dtype) #DataType
```

```python
x=nm.array([41,59,85])
y=nm.array([10,54,26])

print("X + Y",x+y)
print("X - Y",x-y)
print("X * Y",x*y)
print("X / Y",x/y)
```

```python
add=nm.add(x,y)
print("Subtraction",nm.subtract(x,y))
print("Multiplication",nm.multiply(x,y))
print("Division",nm.divide(x,y))
print("Modulo",nm.mod(x,y))
```

```python
a=nm.array([10,20,30,
           40,50,52,36,441])

print("Max Value :",nm.max(a))
print("Min Value :",nm.min(a))
print("Sum Value :",nm.sum(a))
print("Mean Value :",nm.mean(a))#average
```

```python
array1=nm.array([[96,74,63],[85,74,96]])
array2=nm.array([[25,55,41],[96,15,45]])
print("Addition",nm.add(array1,array2))
print("Subtraction",nm.subtract(array1,array2))
print("Modulas",nm.mod(array1,array2))
print("Mean",nm.mean(array1))
print("Mean",nm.mean(array2))
print("Sum",nm.sum(array1))
print("Sum",nm.sum(array2))
```
