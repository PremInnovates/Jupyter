```python
import pandas as pd
StMarks=pd.read_csv('P1.csv')
print(StMarks)
```

```python
new_std=StMarks.dropna() #columns Name Display For use Dropna() Function
print(new_std)
```

```python
fill_std=StMarks.fillna(100.00) #Fill Empty Values
print(fill_std)
```

```python
std1=StMarks["TotalMarks"].fillna(500)
print(std1)
```

```python
StMarks.fillna({"TotalMarks": 600}, inplace=True)
print(StMarks)
```

```python
mean=StMarks["TotalMarks"].mean()
print(mean)
median=StMarks["TotalMarks"].median()
print(median)
mode=StMarks["TotalMarks"].mode()
print(mode)
```
