```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("weatherHistory.csv")
```

```python
print(df.head())
```

```python
print(df.columns)
```

```python
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], errors='coerce')

print(df)
```

```python
df["Temperature (C)"].mean()
```

```python
df[df["Temperature (C)"]>df["Temperature (C)"].mean()]
```

```python
df["Humidity"].max()
```

```python
df["Humidity"].min()
```

```python
weather_counts=df['Precip Type'].value_counts()
print(weather_counts)
```

```python
df['Temp_Diff']=df['Temperature (C)']-df['Apparent Temperature (C)']
print(df['Temp_Diff'])
```

```python
plt.hist(df['Temperature (C)'],color='blue', edgecolor='black')
plt.xlabel('Temperature (C)')
plt.ylabel('Frequency')
plt.title("Histogram of Temperature")
plt.show()
```

```python
plt.plot(df['Formatted Date'], df['Temperature (C)'],marker='o',linestyle='-')
plt.xlabel("Date")
plt.ylabel("Temperature (C)")
plt.title("Temperature Over Time")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
